### Number of genres

```python
#To get the desired result, split the into multicolumns
#For this to work well, remove whitespaces and symbols
updated_df['Genre']=updated_df['Genre'].str.replace(" ", "")
updated_df['Tags']=updated_df['Tags'].str.replace(" ", "")
updated_df['Actors']=updated_df['Actors'].str.replace(" ", "")
updated_df['Title']=updated_df['Title'].str.replace('#',"")
updated_df['Title']=updated_df['Title'].str.replace('+',"")
```

```python
# Create a function
def split_multicolumn(col_series):
    result_df = col_series.to_frame()
    options = []
    for idx, value in col_series[col_series.notnull()].iteritems():
        for option in value.split(','):
            if not option in result_df.columns:
                options.append(option)
                result_df[option] = False
                
            result_df.at[idx, option] = True
            
    return result_df[options]
```

```python
split_genre= split_multicolumn(updated_df.Genre)
split_genre
```

```python
genre_percentage = (split_genre.mean().sort_values(ascending=False)*100).round(2)
genre_percentage
```

```python
plt.figure(figsize=(20,10))
plt.title("The percentage of Genre count")
plt.ylabel('Genres')
plt.xlabel("Percentage[%]")
sns.barplot(x=genre_percentage, y=genre_percentage.index);
```

```python
def count_rows(row):
    return len(row)
```

```python
def heatmap(col1,col2):
    by_rows = updated_df.groupby([col1,col2]).apply(count_rows)
    plt.figure(figsize=(20,10))
    pivot=by_rows.unstack()
    return sns.heatmap(pivot,annot=True)
```
